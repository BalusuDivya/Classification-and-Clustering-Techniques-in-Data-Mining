```python
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt
# Distance matrix provided
points = ['P1', 'P2', 'P3', 'P4', 'P5']
dist_matrix = np.array([
    [1.00, 0.1, 0.41, 0.55, 0.35],
    [0.1, 1.00, 0.64, 0.47, 0.98],
    [0.41, 0.64, 1.00, 0.44, 0.85],
    [0.55, 0.47, 0.44, 1.00, 0.76],
    [0.35, 0.98, 0.85, 0.76, 1.00]])
# Convert the distance matrix to condensed form (upper triangular, without diagonal)
condensed_dist_matrix = [dist_matrix[i][j] for i in range(len(points)) for j in range(i + 1, len(points))]
# Single Link Clustering
single_link = linkage(condensed_dist_matrix, method='single')
# Plot the dendrogram
plt.figure(figsize=(8, 4))
dendrogram(single_link, labels=points, distance_sort='ascending')
plt.title("Single Link Dendrogram")
plt.xlabel("Points")
plt.ylabel("Distance")
plt.show()
# Print each step of cluster combinations
print("Cluster combinations at each step:")
for i, (c1, c2, dist, n_points) in enumerate(single_link):
    print(f"Step {i+1}: Combine clusters {int(c1)+1} and {int(c2)+1} at distance {dist:.2f}")
```
