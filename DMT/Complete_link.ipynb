```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from scipy.spatial.distance import squareform

# Define the distance matrix
distance_matrix = np.array([
    [0.00, 0.10, 0.41, 0.55, 0.35],
    [0.10, 0.00, 0.64, 0.47, 0.98],
    [0.41, 0.64, 0.00, 0.44, 0.85],
    [0.55, 0.47, 0.44, 0.00, 0.76],
    [0.35, 0.98, 0.85, 0.76, 0.00]])

# Create a condensed distance matrix from the square matrix
condensed_distance = squareform(distance_matrix)

# Perform hierarchical clustering using maximum distance (complete linkage)
Z = linkage(condensed_distance, method='complete')

# Plot the dendrogram
plt.figure(figsize=(10, 5))
dendrogram(Z, labels=['P1', 'P2', 'P3', 'P4', 'P5'])
plt.title('Complete Linkage Dendrogram (Maximum Distance)')
plt.xlabel('Points')
plt.ylabel('Distance')
plt.show()

# Print each step of cluster combinations
print("Cluster combinations at each step:")
for i, (c1, c2, dist, n_points) in enumerate(Z):
    print(f"Step {i+1}: Combine clusters {int(c1)+1} and {int(c2)+1} at distance {dist:.2f}")


```
